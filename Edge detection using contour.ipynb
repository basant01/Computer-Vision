```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image=cv2.imread('Pictures/New folder/hands.jpg')
#for copy the image

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python


# Convert to grayscale
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

# Create a binary thresholded image
retval, binary = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)

plt.imshow(binary, cmap='gray')


```

```python
#Find Contours from threshold image
retval,contours,hierarchy=cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

# Draw all Contours in a copy of original image\
image_copy=np.copy(image)
# on which imageto display,which contour to display, -1 means to display all contours,color and size of contour
all_contours= cv2.drawContours(image_copy,contours,-1,(0,255,0),2)

# to display contours
plt.imshow(all_contours)
```
