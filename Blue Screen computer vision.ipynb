```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('Pictures/New folder/pizza_bluescreen.jpg')
```

```python
print(type(image),image.shape)
```

```python
plt.imshow(image)
```

```python
#for copy the image
image_copy = np.copy(image)

#change color BGR to RGB
image_copy=cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
#DEFINE COLOR THERSOLD
```

```python
#Define our color selection boundaries
low=np.array([0,0,230])
high=np.array([50,50,255])
```

```python
#define the mask area
#in range function takes an image on our lower and higher bound values and defines the mask asking if the color value of each pixel
#falls in the range of upper and lower if fall in range mask will print if not it turns pixel into black
mask=cv2.inRange(image_copy,low,high)
```

```python
plt.imshow(mask,cmap='gray')
```

```python
#mask the image to see the pizaa
masked_image=np.copy(image_copy)
masked_image[mask!=0]=[0,0,0]
plt.imshow(masked_image)
```

```python
#first read the image and load in the background and convert it to RGB
image_change = cv2.imread('Pictures/New folder/sky.jpg')

image_change=cv2.cvtColor(image_change,cv2.COLOR_BGR2RGB)
```

```python
# than crop the image to the original size image
crop_background=image_change[0:514,0:816]

crop_background[mask==0]=[0,0,0]
plt.imshow(crop_background)
```

```python
# totalling the both image
total_image = crop_background+masked_image

plt.imshow(total_image)
```
