```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image=cv2.imread('Pictures/New folder/Butterfly.jpg')
#for copy the image
image_copy = np.copy(image)

#change color BGR to RGB
image_copy=cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
gray=cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
# converting it to floating point values for haris detector
gray = np.float32(gray)
#using opencv function haris for corner detector(float value , size of window (here 2 means 2*2  square ),size of sobel operator)
#(a constant value helps to define which points are considered as corners)
dst = cv2.cornerHarris(gray,2,3,0.04)

#dilate corner to see them better
dst =cv2.dilate(dst,None)

plt.imshow(dst,cmap='gray')
```

```python
#define threshold for exactly detecting strong corner
#This value depended on the image
thresh=0.01*dst.max()
# corners to be detected on copy_image
corner_image=np.copy(image_copy)

#iterate through all the corner and detected when threshold is more

for j in range(0,dst.shape[0]):
 for i in range(0,dst.shape[1]):
    if (dst[j,i]>thresh): 
# image , centre point,radius , color,thickness
        cv2.circle(corner_image,(i,j),1,(0,255,255),1)
plt.imshow(corner_image)   
```
